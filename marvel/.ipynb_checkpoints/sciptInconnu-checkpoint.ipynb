```python
from pyspark import SparkContext,SparkConf
conf = SparkConf().setMaster("local").setAppName("Crimes")
sc = SparkContext(conf = conf)
```

```python
def showRdd(rdd):
    for result in rdd.collect():
        print(result)
```

```python
# 
names = sc.textFile("/home/jovyan/work/marvel/Marvel-Names.txt")
showRdd(names)
```

```python
lines = sc.textFile("/home/jovyan/work/marvel/Marvel-Graph.txt")
lines.take(1)
```

```python
showRdd(lines)
```

```python
# ecrire une fonction qui calcule pour chaque superheros le nombre de personnes qu'il connait
# (person_id, count of number of person whom he knows)
# max ()
# au lieu d'avoir person_id ==> person_name
```

```python
# ecrire une fonction qui calcule pour chaque superheros le nombre de personnes qu'il connait
def occurences(line):
    elements = line.split()
    return (elements[0], len(elements)-1)


totalFriendsByCharacter = lines.map(occurences).reduceByKey(lambda x, y : x + y)
showRdd(totalFriendsByCharacter)
# le reduceByKey au cas ou on a :
# (christophe, 87)
# ....
# (christophe, 20)
```

```python
mostPopular = totalFriendsByCharacter.max(lambda x : x[1])
print(mostPopular)
```

```python
def parseNames(line):
    fields = line.split('\"')
    return (int(fields[0]), fields[1].encode("utf8"))
namesRdd = names.map(parseNames)
namesRdd.take(1)
```

```python
mostPopularName = namesRdd.lookup(mostPopular[0])[1]
print(mostPopularName)
```
